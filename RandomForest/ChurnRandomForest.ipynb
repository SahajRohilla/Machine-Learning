```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#convert total charges data type numeric.. so we need to convert " " to NaN
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
#Duplicate check
data.duplicated().sum()
```

```python
#Drop extra column customerID
data.drop(columns='customerID', inplace=True)
```

```python
data.info()
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
col_list = []
for i in data.columns:
    if((data[i].dtype =='object') & (i != 'Churn')):
        col_list.append(i)
```

```python
col_list
```

```python
for i in col_list:
    data[i] = le.fit_transform(data[i])
```

```python
data.info()
```

```python
#Divide the data into x and y
x = data.iloc[:,:-1]
y = data['Churn']
```

```python
#Test train split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify=y)

#Stratify = y-> It will equally divide the values of dependent column between training and testing data
```

```python
#Model Building
from random import Random
from sklearn.ensemble import RandomForestClassifier


#HyperParameters -> Settings(Controls that we choose before training the model)
#n_estimators -> number of trees
#max_depth -> maximum depth of each tree
#min_samples_split -> minimum samples needed to split the data
#min_samples_leaf -> minimum samples in lead
#max_features -> number of features to consider at each split

model1 = RandomForestClassifier(n_estimators=50)
model1.fit(x_train, y_train)
y_pred = model1.predict(x_test)
y_pred 
```

```python
from sklearn.metrics import *
accuracy_score(y_pred, y_test) * 100
```

```python
cm = confusion_matrix(y_pred, y_test)
cm
```

```python
#How to choose optimal value for our hyperparameters
#HYPERPARAMETER Tunning : Choosing the best hyperparameter setting to get the highest accuracy

#Grid Search CV(Cross validation) --> Try all possible combination
#CV -> How many folds you want
#cv = 5(equal dolds/parts)-- model train 4 parts and model test 1 part
```

```python
#Grid search cv
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators' : [100,200,300],
    'max_depth' : [None, 5, 10],
    'min_samples_split' : [2,5,10],
    'min_samples_leaf' : [1,2,3],
    'criterion' : ['gini', 'entropy']
}

model2 = RandomForestClassifier(random_state=4)

grid_search = GridSearchCV(estimator=model2, param_grid=param_grid, cv=2)

grid_search.fit(x_train, y_train)
```
