Logistic Regression
Build a simple Logistic Regression model where target column is 'Churn' and independent variable is 'MonthlyCharges'

(a) - Divide the dataset in 65:35 ratio

(b) - Build the model on train set and predict the values on test set

(c) - Build the confusion matrix and also get the accuracy score


```python
x = data[['MonthlyCharges']]
y = data['Churn']

#Split the data in training and testing 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.35, random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr.fit(x_train, y_train)
```

```python
y_predict = lr.predict(x_test)
```

```python
y_predict
```

```python
#Matrics 

from sklearn.metrics import accuracy_score
print('Accuracy is - ', accuracy_score(y_predict, y_test) * 100)

from sklearn.metrics import confusion_matrix
print('confusion_matrix : ', confusion_matrix(y_test, y_predict))
```
